```python
import keras
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.fashion_mnist.load_data()
```

```python
cl=['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot']
print(cl)
```

```python
for i in range(25):
    plt.figure(figsize=(16,16))
    plt.subplot(5,5,i+1)
    plt.imshow(x_train[i],cmap="Blues")
    plt.axis("off")
    plt.title(cl[y_train[i]]+"="+str(y_train[i]),fontsize=20)
```

Feature Scaling

```python
#basicly this feature scaling is just to scale your data within the range of 0 to 1
#since we have color max value of 255
x_train=x_train/255
x_test=x_test/255

x_train[0]
```

```python
#lets  buit the keras model
model= keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28,28]),
    keras.layers.Dense(units=32,activation='relu'),
    keras.layers.Dense(units=10,activation='softmax')
    
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=1)
#epoch is used to increase the effeciency the data is train multiple time as the number of epoch.
```

```python
model.fit(x_train,y_train,epochs=10)
```

Test and evaluate network.

```python
model.evaluate(x_test,y_test)
```

```python
#since we can see the network is showing the decreace in accuracy by 2% , so we can asume our model in a genralize model and not a overfit model
```

```python
y_predict=model.predict(x_test)
y_predict[0].round(2)
```

```python
np.argmax(y_predict[0].round(2))
#theindex 9 has a ankleboot
```

```python
y_test[0]
#which is actual value, but what we want is image of the prediction
```

```python
for i in range(25):
    plt.figure(figsize=(16,16))
    plt.subplot(5,15,i+1)
    plt.imshow(x_train[i],cmap="Blues")
    plt.axis("off")
    plt.title("Actual={}\nPrediction={}".format(cl[y_test[i]], cl[np.argmax(y_predict[i])]))
```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test,[np.argmax(i) for i in y_predict])
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(cm, annot=True,fmt="d")
```

```python
#classification report
from sklearn.metrics import classification_report
cr= classification_report(y_test, [np.argmax(i) for i in y_predict],target_names=cl)
```
